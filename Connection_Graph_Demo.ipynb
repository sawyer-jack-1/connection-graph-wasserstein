```python
import GraphX as gx
import ConnectionGraphX as cgx
import numpy as np
import networkx as nx
import scipy as sp
```

```python
NODES = 100
EDGES = 200
seed = 5050

a = nx.adjacency_matrix(nx.gnm_random_graph(NODES, EDGES, seed=seed))

# a = np.array([[0, 1, 1], [1, 0, 1], [1, 1, 0]])
g = gx.GraphX(sp.sparse.csr_matrix.toarray(a))

DIM_CONNECTION = 2
h = cgx.ConnectionGraphX(sp.sparse.csr_matrix.toarray(a), DIM_CONNECTION)
```

```python
print(h.nNodes)
print(h.connectionIncidenceMatrix)
print(h.connectionAdjacencyMatrix)
print(h.connectionLaplacianMatrix)
```

```python
h.setEdgeConnection((0, 1), np.array([[0, 1], [-1, 0]]))
```

```python
print(h.connectionIncidenceMatrix)
print(h.connectionAdjacencyMatrix)
print(h.connectionLaplacianMatrix)
h.draw()
print(h.connectionLaplacianMatrixEigenvalues)
print(h.connectionLaplacianMatrixEigenvectors)
```
